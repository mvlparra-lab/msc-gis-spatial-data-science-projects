Para este ejercicio se utilizará el diccionario resultante del ejercicio anterior. 

El objetivo es obtener aquel producto con el precio más caro. Para realizar este ejercicio deberéis declarar una función llamada getExpensiveProduct, la cual tendrá un parámetro. Una vez tengáis definida vuestra función, deberéis imprimir por pantalla lo siguiente: 

print("El producto más caro es: ", <resultado de la función>, "con un precio:", <resultado de la función>)

Donde <resultado de la función> serán el nombre del producto y su precio.

**(1 punto)** 

```python
# Función que devuelve el producto con el precio más alto
def getExpensiveProduct(diccionario):
    producto_caro = None
    precio_max = 0
    
    for producto, precio in diccionario.items():
        if precio > precio_max:
            precio_max = precio
            producto_caro = producto
            
    return producto_caro, precio_max

# Llamamos a la función y mostramos el resultado
producto_mas_caro = getExpensiveProduct(productos)

print("El producto más caro es: ", producto_mas_caro[0], "con un precio:", producto_mas_caro[1])
```

```python
"""
Ejercicio 5.a)

Carga y análisis exploratorio inicial del dataset curso_2019.csv.

El archivo presenta una estructura no estándar, ya que la primera línea
contiene los nombres de las variables precedidos por el carácter "#".
Si se utilizara directamente el parámetro comment="#", se perderían
los nombres originales de las columnas.

Por este motivo, se extrae manualmente la primera línea y se utiliza
como cabecera del DataFrame, garantizando así una correcta estructuración
de los datos para su posterior análisis en la Práctica 2.
"""
# Librería principal para análisis de datos en formato CSV
import pandas as pd

# 1) Leemos la primera línea para extraer los nombres de columnas
with open("P1/P1/data/curso_2019.csv", "r", encoding="utf-8", errors="replace") as f:
    header_line = f.readline().strip()

# Quitamos el prefijo "#Alumnos:" y normalizamos
header_line = header_line.replace("#Alumnos:", "").strip()

# Separamos por ";" para obtener la lista de nombres
columnas = [c.strip() for c in header_line.split(";")]

# 2) Leemos el CSV saltando la primera línea (la cabecera comentada)
df = pd.read_csv(
    "P1/P1/data/curso_2019.csv",
    sep=";",
    skiprows=1,
    names=columnas,
    encoding="utf-8"
)

# 3) Revisión EDA solicitada en el apartado a)
print("Tipos de datos por columna:")
print(df.dtypes)

print("\nNúmero de valores nulos por columna:")
print(df.isnull().sum())
```

```python
"""
Ejercicio 5.b)

Visualización de los nombres de las variables,
el número total de registros y las 5 primeras filas
del DataFrame.
"""

# Mostrar los nombres de las variables
print("Nombres de las variables:")
print(list(df.columns))

# Mostrar número de registros (filas)
print("\nNúmero total de registros:")
print(df.shape[0])

# Mostrar las 5 primeras filas del DataFrame
print("\nPrimeras 5 filas del dataset:")
print(df.head())
```

```python
"""
Ejercicio 5.c)

1) Normalización de los nombres de las variables para evitar
   problemas derivados de tildes, mayúsculas y espacios.
2) Identificación y validación de variables cualitativas dicotómicas.
"""

# --- Normalización de nombres de columnas ---

# Se transforman a minúsculas, se eliminan espacios y tildes
# para garantizar consistencia y facilitar futuras operaciones.
df.columns = (
    df.columns
    .str.lower()
    .str.strip()
    .str.normalize('NFKD')
    .str.encode('ascii', errors='ignore')
    .str.decode('utf-8')
)

print("Nombres de columnas normalizados:")
print(list(df.columns))


# --- Revisión de variable dicotómica ---

"""
Durante el análisis exploratorio se detecta que la variable 'sexo'
presenta inconsistencias en la capitalización ('Femenino' y 'femenino'),
lo que genera una categoría adicional incorrecta.

Se procede a estandarizar los valores para asegurar que la variable
sea estrictamente dicotómica.
"""

df["sexo"] = df["sexo"].str.strip().str.capitalize()

print("\nCategorías finales de la variable 'sexo':")
print(df["sexo"].unique())
print("Número de categorías:", len(df["sexo"].unique()))


# --- Identificación general de variables cualitativas dicotómicas ---

print("\nVariables cualitativas dicotómicas y sus categorías:")

for col in df.select_dtypes(include="object").columns:
    valores = df[col].dropna().unique()
    if len(valores) == 2:
        print(col, ":", valores)
```

```python
"""
Ejercicio 5.d)

Análisis descriptivo de la variable cuantitativa discreta 'edad'.
Se analiza su distribución estadística básica y se calcula
la media utilizando una función distinta a la empleada en la descripción.
"""

# Descripción estadística de la variable edad
print("Descripción estadística de la variable 'edad':")
print(df["edad"].describe().round(3))

# Cálculo de la media utilizando una función diferente
media_edad = df["edad"].mean()

print("\nMedia de la edad:")
print(media_edad)
```

```python
"""
Ejercicio 5.e)

Representación gráfica de la distribución de la variable 'sexo'
mediante un gráfico circular, utilizando el método plot(kind='pie')
según la documentación consultada.
"""

# Recuento por categoría
conteo_sexo = df.groupby("sexo").size()

# Representación gráfica
colores = ["lightcoral", "steelblue"]

conteo_sexo.plot(
    kind="pie",
    autopct="%1.1f%%",
    startangle=90,
    colors=colores
)

import matplotlib.pyplot as plt
plt.ylabel("")  # Elimina etiqueta automática del eje Y
plt.title("Distribución porcentual por sexo")
plt.show()
```

El objetivo es crear una función parametrizada, donde al pasarle una dirección, nos devuelva las coordenadas geográficas (latitud y longitud). A continuación mostramos el orden que debéis seguir para este ejercicio. Tenéis marcado en cada celda el apartado que deberéis desarrollar.

a) Codificad una función parametrizada que, al pasarle un valor (dirección), nos devuelva las coordenadas geográficas.

b) Comprobad que las direcciones del df que habéis analizado sean correctas. Para ello, usad la función que habéis creado en el apartado a) para saber si las direcciones son correctas. Seguid los pasos que se detallan más adelante para la realización de este apartado. 

**(3 puntos)**

```python
# a) 
"""
Ejercicio 6.a)

Función parametrizada para geocodificar una dirección y devolver
coordenadas geográficas (latitud, longitud).
"""

from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter

# Inicializamos el geocodificador (user_agent obligatorio)
geolocator = Nominatim(user_agent="unigis_p1_geocoder")

# Rate limiter para evitar bloqueos por demasiadas peticiones
geocode = RateLimiter(geolocator.geocode, min_delay_seconds=1)

def getCoordinates(direccion):
    """
    Devuelve (latitud, longitud) a partir de una dirección.
    Si no se encuentra, devuelve (None, None).
    """
    if direccion is None or str(direccion).strip() == "":
        return (None, None)
    
    location = geocode(direccion)
    
    if location is None:
        return (None, None)
    
    return (location.latitude, location.longitude)

#Prueba
print(getCoordinates("Girona, España"))
print(getCoordinates(""))
print(getCoordinates(None))
```

Finalmente, tendréis que comprobar que las direcciones que tenéis en la columna 'direccion' sean correctas. A partir de la función que habéis creado anteriormente, deberéis imprimir por pantalla las coordenadas de todas las direcciones. En caso de que haya un error, mostradlo por pantalla junto con las demás coordenadas. Utilizad try/except: https://problemsolvingwithpython.com/08-If-Else-Try-Except/08.05-Try-Except-Statements/

Se puede dar el caso de que la dirección que tengáis almacenada no sea correcta y genere un error. Es fundamental que entendáis de qué error se trata y encontréis una solución. 

**Consejo**: En caso de que encontréis una dirección errónea, podéis eliminar toda la fila del df.

Este ejercicio es importante que lo entendáis bien, ya que representa uno de los mayores retos para los profesionales en SIG. Entender los datos y prepararlos son una de las tareas que más tiempo consumen.

```python
# b)
"""
Ejercicio 6.b)
Comprobación de todas las direcciones usando try/except.
Se imprimen coordenadas o mensaje de error.
"""

for direccion in df["direccion"]:
    
    try:
        lat, lon = getCoordinates(direccion)
        
        if lat is None or lon is None:
            print("NO ENCONTRADA ->", direccion)
        else:
            print("OK ->", direccion, " | ", lat, lon)
    
    except Exception as e:
        print("ERROR ->", direccion, " | ", e)
```

Imprimid ahora las cordenadas sin ningún error.

```python
"""
1) Eliminación de dirección incoherente.
2) Corrección de direcciones con topónimos complejos.
3) Geocodificación final y muestra de coordenadas sin errores.
"""

# 1) Eliminación de la fila incoherente
"""
Depuración de direcciones

Se elimina el registro correspondiente a Madrid por dos motivos:

1) Inconsistencia interna en la dirección (código postal no coherente con la ciudad).
2) Ruptura del patrón espacial del dataset, ya que el resto de registros
   se localizan en Cataluña.

Dado que el objetivo es preparar un conjunto de datos coherente
para análisis posterior, se considera justificada su eliminación.
"""
df = df.drop(index=9)

# 2) Corrección de direcciones no resolubles
df.loc[3, "direccion"] = "Carrer de les Hortènsies, Olot, España"
df.loc[7, "direccion"] = "Teatre Grec, Barcelona, España"

# 3) Geocodificación final e impresión únicamente de coordenadas
for direccion in df["direccion"]:
    try:
        lat, lon = getCoordinates(direccion)
        
        if lat is not None and lon is not None:
            print(round(lat,6), round(lon,6))
    
    except:
        pass
```

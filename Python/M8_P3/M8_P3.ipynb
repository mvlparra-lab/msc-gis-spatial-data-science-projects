```python
# a) y b)

# Se vuelven a abrir las capas para reiniciar los iteradores
obras = fiona.open('data/obras.geojson', 'r', encoding='utf-8', crs='epsg:4326')
distritos = fiona.open('data/distritos-bilbao.geojson', 'r', encoding='utf-8', crs='epsg:3042')

import warnings
warnings.filterwarnings("ignore")

feature_list = []

for distrito in distritos:
    # Convierte la geometría del distrito a un objeto Shapely
    geom_dis = shape(distrito.get('geometry'))
    
    # Inicializa el contador de obras
    count = 0
    
    for obra in obras:
        # Proyección de las coordenadas de la obra a EPSG:3042
        epsg3042 = pyproj.Proj('EPSG:3042')

        geom_metric = epsg3042(
            obra.get('geometry').get('coordinates')[0],
            obra.get('geometry').get('coordinates')[1]
        )
        
        # Crea un punto con las coordenadas proyectadas
        point = Point(geom_metric)

        # Comprueba si la obra está dentro del distrito
        if geom_dis.contains(point):
            count += 1
            
    # Añade el número de obras al distrito
    distrito['properties']['obras'] = count
    feature_list.append(distrito)
    
    # Reabre la capa obras para reiniciar el iterador
    obras = fiona.open('data/obras.geojson', 'r', encoding='utf-8', crs='epsg:4326')
    
schema = {
    'geometry': 'MultiPolygon',
    'properties': {
        'BAR_CL': 'float',
        'BAR_BARR': 'str',
        'BAR_DS_O': 'str',
        'BAR_MUNI': 'str',
        'BAR_MUNI_D': 'str',
        'BAR_PROV': 'str',
        'BAR_PROV_D': 'str',
        'Shape_Leng': 'float',
        'Shape_Area': 'float',
        'distrito': 'int',
        'obras': 'int'
    }
}

with fiona.open('distritos_obras.geojson', 'w', 'GeoJSON', schema=schema, crs=distritos.crs) as layer:
    for distrito in feature_list:
        layer.write(distrito)
```
